<u><b><h3>1.How many are there in each Team and the percentage splitting with respect to the total employees.

```python
team_employee_counts=df['Team'].value_counts()
print("Number of Employees in each team:\n")
print(team_employee_counts)
percentage_split=round((team_employee_counts/len(df))*100,2)
print("\nPercentage spliting:\n")
print(percentage_split)
```

```python
team_table=pd.DataFrame({'Team':team_employee_counts.index, 'Number of Employees':team_employee_counts.values,'Percentage':percentage_split.values})
team_table
```

```python
plt.figure(figsize=(10,6))
plt.bar(team_table['Team'],team_table['Percentage'])
plt.title('Percentage of Employees in each team')
plt.xlabel('Team')
plt.ylabel('Percentage')
plt.xticks(rotation=45 ,ha='right')
plt.show()
```

<b>Data story<b/> - The graph shows a quick overview of percentage of Employees in each team in ABC Company.New Orleans Pelicans has more employees whereas Orlando Magic and Minnesota Timberwolves has the least number of employees

<u><b><h3>3.Find from which age group most of the employees belong to.

```python
df['Age'].value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Age',data=df,hue='Age')
plt.show()
```

```python
bins=[18,20,25,30,35,40]
labels=['18-20','20-25','25-30','30-35','35-40']
df['age_groups']=pd.cut(df['Age'],bins=bins,labels=labels,right=True)
df
```

```python
age_groups=age_groups.value_counts()
age_groups
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='age_groups',data=df,hue='age_groups')
plt.show()
```

<b>Data Story- As shown in the above graph.We can clearly see that the number of employees are more in the age category of 20-25 with 179 employees and 165 employees in the age category of 25-30.<b/> 

<u><b><h3>4.Find out under which team and position, spending in terms of salary is high.


```python
high_spending=df.groupby(['Team','Position'])['Salary'].sum().idxmax()
high_spending
```

<b>The team with the highest spending in terms of salary is 'Los Angeles Lakers' and the position is 'SF'<b/>

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Team',y='Salary',hue='Position',data=df[df['Team']== high_spending[0]])
plt.title('Salary Distribution in High Spending Team and Position')
plt.xlabel('Team')
plt.ylabel('Salary')
plt.show()
```

<b>Data Story-As shown in the above graph,we can clearly see that SF position in Los Angeles Lakers Team has the high spending in salary distribution.

<u><b><h3>5.Find if there is any correlation between age and salary , represent it visually.


```python
correlation=df[['Age','Salary']].corr()
correlation
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(correlation,annot=True,cmap='coolwarm',fmt=".2f", linewidths=.5)
plt.title('Correlation between Age and Salary')
plt.show()
```

<b>Data Story - The heatmap represents the correlation matrix between Age and Salary
